## Manual Selection of Features
Through a process of testing different combinations of features on the models, we decided to drop the following features to get the best scores.

```python
# configs to drop entire groups of features
DROP_WEATHER_CODES = False
DROP_LOCATIONS = True
DROP_ROLLING_WEATHER = True

DROP_TIMESTAMP = True
DROP_LOC_INTERACTION = False
```

```python
# Get the columnsnames
cols = list(X_train.columns)
# List of columns to drop
to_drop = []

if DROP_WEATHER_CODES:
    # all weather Codes
    to_drop += [i for i in cols if 'code_' in i]

if DROP_LOCATIONS:
    # all locations
    to_drop += [i for i in cols if 'loc_' in i] 

if DROP_ROLLING_WEATHER:
    # all rolling weather data
    for s in ['_14d', '_30d', '_90d']:
        to_drop += [i for i in cols if s in i]

if DROP_TIMESTAMP:
    to_drop += ['timestamp']

if DROP_LOC_INTERACTION:
    to_drop += ['IntLoc']
        
X_train.drop(columns=to_drop, inplace=True)
X_val.drop(columns=to_drop, inplace=True)
X_test.drop(columns=to_drop, inplace=True)

list(X_train.columns)
```

### Recursive Feature Elimination
This is a technique of recursively running a selected estimator and dropping features based on the lowest scoring feature by coef or feature importance. The process is repeated until the specified number of features are attained in RFE or in the case of RFECV it is cross validated and the number of features are selected automatically.

```python
X_train = pd.read_csv('../datasets/preprocessed_datasets/x_train.csv')
X_val = pd.read_csv('../datasets/preprocessed_datasets/x_val.csv')
y_train = pd.read_csv('../datasets/preprocessed_datasets/y_train.csv')
y_val = pd.read_csv('../datasets/preprocessed_datasets/y_val.csv')
X_test = pd.read_csv('../datasets/preprocessed_datasets/x_test.csv')

y_train = y_train['WnvPresent']
y_val = y_val['WnvPresent']
```

```python
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
```

#### Recursive Feature Elimination using LinearSVC

```python
%%time
estimator = LinearSVC(max_iter=1000, C=0.01)
rfe_svc = RFE(estimator, n_features_to_select=77, step= 1)
rfe_svc = rfe_svc.fit(X_train_ss, y_train)
```

#### Recursive Feature Elimination using DecisionTreeClassifier

```python
%%time
estimator = DecisionTreeClassifier()
rfe_dtc = RFE(estimator, n_features_to_select=77, step= 1)
rfe_dtc = rfe_dtc.fit(X_train, y_train)
```

#### Cross Validated Recursive Feature Elimination using DecisionTreeClassifier

```python
%%time
estimator = DecisionTreeClassifier()
rfecv_dtc = RFECV(estimator, step=1, cv=5)
rfecv_dtc = rfecv_dtc.fit(X_train, y_train)
```

#### Cross Validated Recursive Feature Elimination using LinearSVC

```python
%%time
estimator = LinearSVC(max_iter=1000, C=0.01)
rfecv_svc = RFECV(estimator, step=1, cv=5)
rfecv_svc = rfecv_svc.fit(X_train_ss, y_train)
```

### Principal Component Analysis
Performant for datasets where features have large ranges in values. Not likely to produce good results as our dataset has many encoded categories with 0-1 range.

```python
# Load the data
X_train = pd.read_csv('../datasets/preprocessed_datasets/x_train.csv')
X_val = pd.read_csv('../datasets/preprocessed_datasets/x_val.csv')
y_train = pd.read_csv('../datasets/preprocessed_datasets/y_train.csv')
y_val = pd.read_csv('../datasets/preprocessed_datasets/y_val.csv')
X_test = pd.read_csv('../datasets/preprocessed_datasets/x_test.csv')

y_train = y_train['WnvPresent']
y_val = y_val['WnvPresent']
```

```python
# Scale the data with standard scaler
ss = StandardScaler()
ss.fit(X_train)
X_train_ss = ss.transform(X_train)
X_val_ss = ss.transform(X_val)
X_test_ss = ss.transform(X_test)
```

```python
# Perform PCA Transformation
pca = PCA()
pca.fit(X_train_ss)
Z_train = pca.transform(X_train_ss)
Z_val = pca.transform(X_val_ss)
Z_test = pca.transform(X_test_ss)
```

```python
pca_feature_rank = pd.DataFrame(pca.components_,columns=X_train.columns).mean().abs().sort_values(ascending = False)
```

```python
pca_feature_rank
```

```python
# Explained variance
var_exp = pca.explained_variance_ratio_
# Cumulative explained variance
cum_var_exp = np.cumsum(var_exp)

# Plot the variance explained (and cumulative variance explained).

# Set figure size.
plt.figure(figsize=(12,8))

# Plot the explained variance.
plt.plot(range(len(var_exp)), var_exp, lw=3, label = 'Variance Explained')

# Plot the cumulative explained variance.
plt.plot(range(len(var_exp)), cum_var_exp, lw=3, color = 'orange', label = 'Cumulative Variance Explained')

# Add horizontal lines at y=0 and y=1.
plt.axhline(y=0, linewidth=1, color='grey', ls='dashed')
plt.axhline(y=.5, linewidth=1, color='grey', ls='dashed')

# Set the limits of the axes.
plt.xlim([-1,21])
plt.ylim([-0.01,.51])

# Label the axes.
plt.ylabel('Variance Explained', fontsize=20)
plt.xlabel('Principal Component', fontsize=20)

# Make the tick labels bigger
plt.xticks(range(0, 21, 5), range(1, 22, 5), fontsize=12)
plt.yticks(fontsize=12)
    
# Add title and legend.
plt.title('Component vs. Variance Explained', fontsize=24)
plt.legend(fontsize=11);
```

### SelectKBest
This is a technique of selecting K (number) of best features determined by a classifier. In this case we will use ANOVA F value to determine the best features based on the ratio of mean squares.

```python
# Load the data
X_train = pd.read_csv('../datasets/preprocessed_datasets/x_train.csv')
X_val = pd.read_csv('../datasets/preprocessed_datasets/x_val.csv')
y_train = pd.read_csv('../datasets/preprocessed_datasets/y_train.csv')
y_val = pd.read_csv('../datasets/preprocessed_datasets/y_val.csv')
X_test = pd.read_csv('../datasets/preprocessed_datasets/x_test.csv')

y_train = y_train['WnvPresent']
y_val = y_val['WnvPresent']
```

```python
# Drop all zero columns

#For each of the X sets, find any features that are all 0 or 1
X_ar = [X_train, X_val, X_test]
zero_cols = []
for x in X_ar:
    zero_cols.append(list(x.loc[:, (x == 0).all(axis=0)].columns))
    zero_cols.append(list(x.loc[:, (x == 1).all(axis=0)].columns))
zero_cols = [item for sublist in zero_cols for item in sublist]
zero_cols = list(set(zero_cols))

# drop those columns
X_train.drop(columns=zero_cols, inplace=True)
X_val.drop(columns=zero_cols, inplace=True)
X_test.drop(columns=zero_cols, inplace=True)
```

```python
# Select the best features using ANOVA F-value
fs = SelectKBest(f_classif, k=77).fit(X_train, y_train)
X_train_fs = fs.transform(X_train)
X_val_fs = fs.transform(X_val)

# sort the scores for observations
sorted_scores = sorted(list(zip(X_train.columns, fs.scores_)), key=lambda x:x[1], reverse=True)

# tabulate the results for printing
total = sum([v for k,v in sorted_scores])
table=[]
cum = 0
count= 0
for k,v in sorted_scores:
    cum += v
    count+=1
    table.append([count, k, round(cum/total*100,1), round(v,4)])
    
# plot the graph
plt.figure(figsize=(12,8))
plt.bar([i for i in range(len(sorted_scores))], [v for k,v in sorted_scores])

# Label the axes.
plt.ylabel('ANOVA F value', fontsize=20)
plt.xlabel('Ranking', fontsize=20)

# Title
plt.title('Feature ranking using ANOVA F value', fontsize=24)
plt.show()

#print the results
print(tabulate(table, headers=['rank', 'feature', 'cumulative%', 'weight']))
```
