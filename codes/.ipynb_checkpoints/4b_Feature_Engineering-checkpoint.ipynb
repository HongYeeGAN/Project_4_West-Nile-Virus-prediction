## Feature Engineering

### Create unix timestamp
unix timestamp was created to describe a linear time relationship

```python
'''
Description:
Converets ordinal dates to continuous unix time
---
Params:

- df: (pandas.DataFrame) dataframe to populate

---
Returns:

- (pandas.DataFrame) df with df['timestamp'] = unix time
'''
def create_unix_timestamp(df):
    dts = [dt.datetime(df.loc[i,'Year'], df.loc[i,'Month'], df.loc[i,'Day']) for i in df.index]
    df['timestamp'] = [time.mktime(dt.timetuple()) for dt in dts]
    return df
```

```python
# Create a unix timestamp from date
train_data = create_unix_timestamp(train_data)
test_data = create_unix_timestamp(test_data)
```

```python
train_data.loc[0, 'timestamp']
```

### Create Lat Long Interaction Term
A lat long interaction term was created to decribe locations in regression modeling

```python
# lat*lng = latitude * longitude
train_data['IntLoc'] = train_data['Latitude'] * train_data['Longitude'] * -1
test_data['IntLoc'] = test_data['Latitude'] * test_data['Longitude'] * -1
```

### categorize codesum
Codesum is the the weather codes of every trap location. They are space separated in the data. Instead of having cumulated weather codes as a category, each location will have code_xx = 1 for each of the codes present in that entry. ie, "BCFG BR" will be BCFG=1 BR=1

```python
'''
Description:
Checks all combinations of codesum and creates a set of possible codes
---
Params:

- df0, df1: (pandas.DataFrame) dataframe to check

---
Returns:

- (str set) set of codes
'''
def categorize_codesum(df0 = train_data, df1 = test_data):
    code_l = np.concatenate((df0['CodeSum'].unique(), df1['CodeSum'].unique()), axis=0)
    nested_l = [d.split() for d in code_l]
    flatten = lambda l: [item for sublist in l for item in sublist]
    return set(flatten(nested_l))

codesum_set = categorize_codesum()
```

```python
codesum_set
```

```python
'''
Description:
Populates dataframe with OHE of categories.
'A B' creates: A=1 B=1 instead of AB=1
---
Params:

- df: (pandas.DataFrame) dataframe to check
- col: (str) column to encode
- cats: (str set) set of codes to check for
- prefix: (str) text to prepend to column label
---
Returns:

- (pandas.DataFrame) populated dataframe with all the new OHE columns
'''
def ohe_space_seperated_cats(df:object, col:str='CodeSum', cats:set=codesum_set, prefix:str='code_'):
    result = []
    for cats_string in df[col]:
        positive_cat_list = cats_string.split()
        row = [1 if cat in positive_cat_list else 0 for cat in cats ]
        result.append(row)
    return pd.DataFrame(result, columns=[f"{prefix}{cat.lower()}" for cat in cats])
```

```python
# Label Encode codesum
train_data = train_data.join(ohe_space_seperated_cats(train_data))
test_data = test_data.join(ohe_space_seperated_cats(test_data))
```

```python
train_data.loc[0, [c for c in list(train_data.columns) if 'code_' in c]]
```

### cluster locations
cluster locations based on distance, calculated by DBSCAN's epsilon

```python
# check distribution to decide between MinMaxScalar or StandardScalar 
lat = train_data['Latitude']
lat.plot.hist(bins=20, alpha=0.5);
```

```python
# check distribution to decide between MinMaxScalar or StandardScalar 
lng = train_data['Longitude']
lng.plot.hist(bins=20, alpha=0.5);
```

Since the distributions are not normal, we will be using MinMaxScale instead of StandardScale

```python
# get all the lat long of train and test set to decide on clusters
train_loc_df = train_data[['Latitude', 'Longitude']].copy()
test_loc_df = test_data[['Latitude', 'Longitude']].copy()

all_locations_df = train_loc_df.append(test_loc_df)
all_locations_df.drop_duplicates(inplace=True)
```

```python
# scale with min max as distribution is not normal
scaler = MinMaxScaler(feature_range = (0, 1))
cluster_features = ['Latitude', 'Longitude']
mms_data = scaler.fit_transform(all_locations_df[cluster_features])

# use dbscan to cluster the locations
dbscan = DBSCAN(
    eps=0.02, # ADJUST THIS FOR DIFFERENT GROUP SIZE
    min_samples=1, # 1 means no outliers. everyone can be an island
    n_jobs=-1,
)
dbscan.fit(mms_data)

# results
print('silhouette_score: ',silhouette_score(mms_data, dbscan.labels_))
print('n outliers: ', list(dbscan.labels_).count(-1))
print('n clusters: ', len(set(dbscan.labels_)))

plt.figure(figsize=(8, 8))
sns.scatterplot(
    data=all_locations_df, 
    y='Latitude', 
    x='Longitude', 
    hue=dbscan.labels_,
    palette='bright',
    legend=False,
);
```

```python
# label clusters to lat long values like a dictionary
all_locations_df['loc'] = dbscan.labels_

# reset the index
all_locations_df.reset_index(drop=True, inplace=True)

# create a string tag to easily map to train test dfs
all_locations_df['loc_tag'] = [f"{np.around(all_locations_df.loc[i,'Latitude'],decimals=6)}{np.around(all_locations_df.loc[i,'Longitude'], decimals=6)}" 
                               for i in all_locations_df.index]

# create a tag: location dict for mapping
loc_dict = {str(all_locations_df.loc[i,'loc_tag']):str(int(all_locations_df.loc[i,'loc'])) 
            for i in all_locations_df.index}

# map loc cluster to train data
train_data['loc'] = train_data['Latitude'].round(decimals=6).astype(str) + train_data['Longitude'].round(decimals=6).astype(str)
train_data['loc'] = train_data['loc'].map(lambda x: loc_dict[x])

# map loc cluster to test data
test_data['loc'] = test_data['Latitude'].round(decimals=6).astype(str) + test_data['Longitude'].round(decimals=6).astype(str)
test_data['loc'] = test_data['loc'].map(lambda x: loc_dict[x])
```

```python
# Encode location clusters
ohe_loc_train = pd.get_dummies(train_data['loc'], prefix='loc')
train_data = train_data.join(ohe_loc_train)
ohe_loc_test = pd.get_dummies(test_data['loc'], prefix='loc')
test_data = test_data.join(ohe_loc_test)
```

### categorize species
Encode the species described in an entry. Test has an additional category "UNSPECIFIED CULEX" which means it could be any of the other species present. We will set it to 1 for all train data to match the presence of that category in test data.

```python
# OHE mosquito species
ohe_species_train = pd.get_dummies(train_data['Species'], prefix='species')
train_data = train_data.join(ohe_species_train)
train_data['species_UNSPECIFIED CULEX'] = 1
```

```python
# OHE mosquito species
ohe_species_test = pd.get_dummies(test_data['Species'], prefix='species')
test_data = test_data.join(ohe_species_test)
```

### drop unnecessary columns
Remove features that have either been processed, irrelevant or unresolvable.

```python
cols = [
    'Unnamed: 0',
    'Date',
    'Address',
    'Species',
    'Block',
    'Street',
    'Trap',
    'AddressNumberAndStreet',
    'NumMosquitos',
    'Station',
    'CodeSum',
    'geometry',
    'AddressAccuracy',
    'loc',
    'Longitude',
    'Latitude',
]
train_data.drop(columns=cols, inplace=True)
```

```python
cols = [
    'Unnamed: 0',
    'Id',
    'Date',
    'Address',
    'Species',
    'Block',
    'Street',
    'Trap',
    'AddressNumberAndStreet',
    'Station',
    'CodeSum',
    'AddressAccuracy',
    'loc',
    'Longitude',
    'Latitude',
]
test_data.drop(columns=cols, inplace=True)
```

### 0 fill missing columns
Fill in missing features in train set. This is a result of location clusters present in test set that is not present in train set

```python
# cols missing from test data
[col for col in train_data.columns if col not in test_data.columns]
```

```python
# cols missing from train data
[col for col in test_data.columns if col not in train_data.columns]
```

```python
# cols missing from train data
missing_col_train = [col for col in test_data.columns if col not in train_data.columns]
for c in missing_col_train:
    train_data[c] = 0
```

```python
# rearrange columns
cols = list(test_data.columns)
cols.append('WnvPresent')
train_data = train_data[cols]
```

### Train Test Split and SMOTE

```python
# X = all features except target
X = train_data.drop(columns=['WnvPresent'], inplace=False)
# Y is target
Y = train_data['WnvPresent']

# shuffle and stratify to minimize effect of imbalanced class
X_train, X_val, y_train, y_val = train_test_split(
    X, 
    Y,
    test_size=.2,
    random_state=42,
    shuffle=True,
    stratify=Y)

# Populate only the train data to address imbalance class for model training
sm = SMOTE(random_state=42, n_jobs=-1, k_neighbors=10)
X_train, y_train = sm.fit_sample(X_train, y_train)
```
