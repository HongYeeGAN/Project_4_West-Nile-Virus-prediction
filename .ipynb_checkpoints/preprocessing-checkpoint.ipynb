### categorize codesum

```python
def categorize_codesum(df0 = train_data, df1 = test_data):
    code_l = np.concatenate((df0['CodeSum'].unique(), df1['CodeSum'].unique()), axis=0)
    nested_l = [d.split() for d in code_l]
    flatten = lambda l: [item for sublist in l for item in sublist]
    return set(flatten(nested_l))
codesum_set = categorize_codesum()
```

```python
codesum_set
```

```python
def ohe_space_seperated_cats(df, col='CodeSum', cats=codesum_set, prefix='code_'):
    result = []
    for cats_string in df[col]:
        positive_cat_list = cats_string.split()
        row = [1 if cat in positive_cat_list else 0 for cat in cats ]
        result.append(row)
    return pd.DataFrame(result, columns=[f"{prefix}{cat.lower()}" for cat in cats])
```

```python
train_data = train_data.join(ohe_space_seperated_cats(train_data))
```

```python
test_data = test_data.join(ohe_space_seperated_cats(test_data))
```

### cluster locations

```python
location_df = train_data['Latitude']
location_df.plot.hist(bins=20, alpha=0.5)
```

```python
location_df = train_data['Longitude']
location_df.plot.hist(bins=20, alpha=0.5)
```

```python
train_loc_df = train_data[['Latitude', 'Longitude']].copy()
test_loc_df = test_data[['Latitude', 'Longitude']].copy()

all_locations_df = train_loc_df.append(test_loc_df)
all_locations_df.drop_duplicates(inplace=True)
```

```python
all_locations_df.head()
```

```python
# scale with min max as distribution is not normal
scaler = MinMaxScaler(feature_range = (0, 1))
cluster_features = ['Latitude', 'Longitude']
mms_data = scaler.fit_transform(all_locations_df[cluster_features])

# use dbscan to cluster the locations
dbscan = DBSCAN(
    eps=0.035,
    min_samples=1,
    n_jobs=-1,
)
dbscan.fit(mms_data)

# results
print('silhouette_score: ',silhouette_score(mms_data, dbscan.labels_))
print('n outliers: ', list(dbscan.labels_).count(-1))
print('n clusters: ', len(set(dbscan.labels_)))

plt.figure(figsize=(8, 8))
sns.scatterplot(
    data=all_locations_df, 
    y='Latitude', 
    x='Longitude', 
    hue=dbscan.labels_,
    palette='bright',
    legend=False,
);
```

```python
# label clusters to lat long values like a dictionary
all_locations_df['loc'] = dbscan.labels_
```

```python
all_locations_df.reset_index(drop=True, inplace=True)
```

```python
all_locations_df['loc_tag'] = [f"{np.around(all_locations_df.loc[i,'Latitude'],decimals=6)}{np.around(all_locations_df.loc[i,'Longitude'], decimals=6)}" for i in all_locations_df.index]
```

```python
loc_dict = {str(all_locations_df.loc[i,'loc_tag']):str(int(all_locations_df.loc[i,'loc'])) for i in all_locations_df.index}
```

```python
train_data['loc'] = train_data['Latitude'].round(decimals=6).astype(str) + train_data['Longitude'].round(decimals=6).astype(str)
train_data['loc'] = train_data['loc'].map(lambda x: loc_dict[x])
```

```python
test_data['loc'] = test_data['Latitude'].round(decimals=6).astype(str) + test_data['Longitude'].round(decimals=6).astype(str)
test_data['loc'] = test_data['loc'].map(lambda x: loc_dict[x])
```

```python
ohe_loc_train = pd.get_dummies(train_data['loc'], prefix='loc')
train_data = train_data.join(ohe_loc_train)
ohe_loc_test = pd.get_dummies(test_data['loc'], prefix='loc')
test_data = test_data.join(ohe_loc_test)
```
